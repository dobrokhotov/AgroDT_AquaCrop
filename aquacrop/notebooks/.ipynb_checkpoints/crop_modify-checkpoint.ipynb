```python
import pandas as pd
from collections import defaultdict
```

```python
cropfile = 'Canola.CRO'
```

```python
with open(cropfile) as f:
    lines = f.readlines()
```

```python
values = [i.strip().split(':',1)[0].strip() for i in lines[1:]]
parameters = [i.strip().split(':',1)[1].strip() for i in lines[1:]]
```

```python
ser = pd.Series(data=values, index=parameters)
```

```python
y = {'Maximum canopy cover (CCx) in fraction soil cover': 0.70}
```

```python
ser.update(y)
```

```python
ser['Maximum canopy cover (CCx) in fraction soil cover']
```

```python
for k, v in ser.items():
    print(k)
    print(v) 
```

```python
project_name = 'Menkovo_2023'
out_cropfile = project_name + '_' + cropfile
with open(out_cropfile, 'w') as f:
    f.write(f'{project_name}\n')
    for k, v in ser.items():
        v_modify = '  ' + str(v)
        v_modify = v_modify + ' '*(15-len(v_modify))
        f.write(f'{v_modify}: {k}\n')
```
